# Data Exploratory Analysis for Data Mining

Download the following data




    https://archive.ics.uci.edu/ml/datasets/Adult


```python
import numpy as np
import pandas as pd
import matplotlib_inline
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

missing_values = [" ?"]
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,na_values = missing_values)
data.columns = ['age', 'workclass', 'fnlwgt', 'education','education-num',
    'marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','class']

data.head(15)
```

# Summary Statistics

For each continuous attribute, calculate its average, standard deviation, minimum, and maximum values.

```python
from pandas.api.types import is_numeric_dtype
for col in data.columns:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Mean = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())
```

For the discrete attribute, count the frequency for each of its distinct values.

```python
for col in data.columns:
    if not is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print(data[col].value_counts())
```

# Data Cleaning

Check for missing values in each column

```python
data.isnull().sum()
```

```python
#function that plots histograms of missing attributes
def plot_na(dataFrame):
    fig = plt.figure(figsize=(27,7))

    fig.add_subplot(3, 1, 1)
    work_hist = dataFrame['workclass'].hist()

    fig.add_subplot(3, 1, 2)
    occu_hist = dataFrame['occupation'].hist()

    fig.add_subplot(3, 1, 3)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=90)
    occu_hist = dataFrame['native-country'].hist()

#histogram before replacing
plot_na(data)
```

Replace missing values with median or mode of the attribute

```python
df1 = data.copy()
print(df1['occupation'].value_counts())
for i in df1.columns:
    df1[i].fillna(df1[i].mode()[0], inplace=True)
    print(df1[i].mode()[0])

print(df1.isnull().sum())
print(df1['occupation'].value_counts())
plot_na(df1)
```

Replace missing values with median or mode of the attribute in the particular class to which the instance belongs  

```python
def replace_na(data, column_name):
    #filter workclass based on class then find mode
    class_1 = data.loc[data['class'] == " <=50K", column_name].mode()[0]
    class_2 = data.loc[data['class'] == " >50K", column_name].mode()[0]

    #print(data[column_name].isnull().sum())
    for index, row in data.iterrows():
        if pd.isnull(row[column_name]):
            if row['class'] ==" <=50K":
                data.loc[index, column_name] = class_1
            else:
                data.loc[index, column_name] = class_2
    #print(data['workclass'].isnull().sum())
    return data
print("Before")
print(data.isnull().sum())
replace_na(data,'workclass')
replace_na(data,'occupation')
replace_na(data,'native-country')

print()
print("After")
print(data.isnull().sum())
plot_na(data)
```
